```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500,500), np.uint8)

cv2.imshow("image", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500,500), np.uint8)

cv2.imshow("image", img)

plt.hist(img.ravel(), 256, [0, 256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500,500), np.uint8)
cv2.rectangle(img, (0,250), (500,500), (255), -1)

cv2.imshow("image", img)

plt.hist(img.ravel(), 256, [0, 256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500,500), np.uint8)
cv2.rectangle(img, (0,250), (500,500), (255), -1)
cv2.rectangle(img, (0,125), (250,250), (127), -1)


cv2.imshow("image", img)

plt.hist(img.ravel(), 256, [0, 256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("lena.jpg")
#img = np.zeros((500,500), np.uint8)
#cv2.rectangle(img, (0,250), (500,500), (255), -1)
#cv2.rectangle(img, (0,125), (250,250), (127), -1)

b,g,r = cv2.split(img)

cv2.imshow("image", img)
cv2.imshow("b", b)
cv2.imshow("g", g)
cv2.imshow("r", r)



plt.hist(img.ravel(), 256, [0, 256])
plt.hist(b.ravel(), 256, [0, 256])
plt.hist(g.ravel(), 256, [0, 256])
plt.hist(r.ravel(), 256, [0, 256])

plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
