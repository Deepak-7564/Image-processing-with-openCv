```python
import numpy as np
import cv2 as cv

def color(x):
    print(x)

#create black image
img = np.zeros((300,512,3), np.uint8)
cv.namedWindow('image')

cv.createTrackbar('Blue','image',0 , 255, color)
cv.createTrackbar('Green','image',0 , 255, color)
cv.createTrackbar('Red','image',0 , 255, color)


while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == 27:
        break
        
    b = cv.getTrackbarPos('Blue','image')
    g = cv.getTrackbarPos('Green','image')
    r = cv.getTrackbarPos('Red','image')
    
    img[:] = [b,g,r]

cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv

def color(x):
    print(x)

#create black image
img = np.zeros((300,512,3), np.uint8)
cv.namedWindow('image')

cv.createTrackbar('Blue','image',0 , 255, color)
cv.createTrackbar('Green','image',0 , 255, color)
cv.createTrackbar('Red','image',0 , 255, color)

switch = '0: OFF\n 1 : ON'
cv.createTrackbar(switch, 'image', 0, 1, color)


while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == 27:
        break
        
    b = cv.getTrackbarPos('Blue','image')
    g = cv.getTrackbarPos('Green','image')
    r = cv.getTrackbarPos('Red','image')
    s = cv.getTrackbarPos(switch,'image')
    
    if s == 0:
        img[:] = 0
    else:
        img[:] = [b,g,r]
        

cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv

def color(x):
    print(x)

#create black image
#img = np.zeros((300,512,3), np.uint8)

cv.namedWindow('image')

cv.createTrackbar('CP','image',10 , 400, color)


switch = 'color/gray'
cv.createTrackbar(switch, 'image', 0, 1, color)


while(1):
    img = cv.imread('lena.jpg')
    pos = cv.getTrackbarPos('CP', 'image')
    font = cv.FONT_HERSHEY_SIMPLEX
    cv.putText(img, str(pos), (50,150), font, 6, (0,0,255),4)
    
    k = cv.waitKey(1) & 0xFF
    if k == 27:
        break
        
    s = cv.getTrackbarPos(switch,'image')
    
    if s == 0:
        pass
    else:
        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    cv.imshow('image',img)
cv.destroyAllWindows()
```
